```python
def neighborhood(grid, location):

    '''
    Extracts the neighborhood of a location and uses a neighborhood extension of one

    Inputs:
        - city grid (list of lists): total grid where the neighborhood (subgrid) is extracted from
        - location (list of two int):
            the row and column index (stating at zero) of the location we extract the neighborhood for.
            examples: [0, 4] or [5, 1]
    Output:
        - neighborhood (list of lists): the neighborhood of this location at the specified extension
    '''
    i, j = location
    N = len(grid)
    subgrid = []

    # Neighborhood indices limits
    up = max(0, i - 1)
    down = min(N, i + 1)
    left = max(0, j - 1)
    right = min(N, j + 1)

    # Extracting neighborhood
    for row in grid[up: down + 1]:
        subgrid.append(row[left: right + 1])

    return subgrid
```

```python
def neighborhood(grid, location):
    
    '''
    Extracts the neighborhood of a location and uses a neighborhood extension of one
    
    Inputs:
        - city grid (list of lists): total grid where the neighborhood (subgrid) is extracted from
        - location (list of two int): 
            the row and column index (stating at zero) of the location we extract the neighborhood for.
            examples: [0, 4] or [5, 1]
    Output:
        - neighborhood (list of lists): the neighborhood of this location at the specified extension
    '''
    
    #initialize neighborhood list
    nh = []
    
    # get the location of the agent
    agent_row = location[0]
    agent_col = location[1]

    # find the max and min row and column index allowed
    # for a lot to be inlcuded in the the neighborhood
    nh_row_max = agent_row + 1
    nh_row_min = agent_row - 1
    nh_col_max = agent_col + 1
    nh_col_min = agent_col - 1

    # Loop over all rows 
    for row_index, row in enumerate(grid) :
        
        # Test if this row is within the max and min 
        # row index for the neighborhood
        if nh_row_min <= row_index and row_index <= nh_row_max :
            
            # If this row is within max min index, 
            # create an empty list for this row, where
            # lots withing the allowed columns indexes
            # will be stored
            nh_row = []
            
            # loop over al lots in this row
            for col_index, col in enumerate(row) :
                
                # Test if this lot is within the 
                # max and min column index
                if nh_col_min <= col_index and col_index <= nh_col_max :
                    
                    # Add lot to neighborhood row
                    nh_row.append(col)
            
            # Add neighborhood row to neighborhood
            nh.append(nh_row)


    # return the neighborhood
    return nh
```

```python
def similarity(grid, location):
    
    '''
    Calculates the similarity score of the location a city for a specified neighborhood location
    Important: remember to use your function neighborhood() in this function
    
    Inputs:
        - city grid (list of lists): total grid
        - location (list of two int): 
            the row and column index (stating at zero) of the location we calculate the similarity score for
    Output:
        - similarity score (float)
    '''
    
    i, j = location
    my_neighborhood = neighborhood(grid, location)
    group = grid[i][j]
    numerator = 0
    denominator = 0

    # Adding units to numerator or denominator
    for street in my_neighborhood:
        for lot in street:
            if lot == group:
                numerator += 1
            if lot != '':
                denominator +=1

    return numerator / denominator
```

## Task 2: Evaluate the hypothetical similarity score of an agent if they moved to another location



```python
def hypothetical_similarity(grid, current_location, new_location):
    
    '''
    Calculates the similarity score an agent would have in a hypothetical location in the city
    
    Inputs:
        - city grid (list of lists): total grid
        - current_location (list of two int): the row and column index (stating at zero) of the current location of the agent
        - new_location (list of two int): the row and column index (stating at zero) where the agent would be moving
    Output:
        - similarity score (float)
    '''
    
    i, j = current_location
    group = grid[i][j]
    i_new, j_new = new_location

    # Creating a copy of the city grid -- that's where we work
    city_clone = clone_grid(grid)
    city_clone[i_new][j_new] = group
    city_clone[i][j] = ''

    return similarity(city_clone, new_location)
```

## Task 3: List all the available and acceptable locations to move for an agent and estimate Euclidean distances


```python
def list_available(grid, location, threshold):
    
    '''
    Creates a list of all the available locations for an agent to move, where their similarity
    score would be at least equal than their acceptable threshold.
    
    Inputs:
        - city grid (list of lists): total grid
        - location (list of two int): current location of the agent who would move
        - threshold (float): minimum similarity accepted by the agent
    Output:
        - list of available locations that meet the minimum similarity score.
            example: [[4, 3], [0, 2], [6, 6], [0, 8]]
    '''
    
    available = []

    for i, street in enumerate(grid):
        for j, lot in enumerate(street):
            if lot == '': # lot is available
                similarity = hypothetical_similarity(grid, location, [i, j])
                if similarity >= threshold:
                    available.append([i, j])

    return available
```

```python
def euclidean_distance(current_location, new_location):
    
    '''
    Calculates the Euclidean distance between two locations
    
    Inputs:
        - current_location (list of two int): list of row and column indices
        - new_location (list of two int): list of row and column indices
    Output:
        - Euclidean distance (float)
    '''
    
    dist = ((current_location[0]-new_location[0])**2 + (current_location[1]-new_location[1])**2)**(1/2)

    return dist
```

## Task 4: Move an agent to another location if needed and update the city grid

```python
def move_agent(grid, location, threshold):
    
    '''
    Evaluates if the agent in location is dissatisfied. If they are, the agent is moved to another
    location using the moving rule and the grid is updated.
    
    Inputs:
        - city grid (list of lists): total grid
        - location (list of two int): current location of the agent who is evaluated
        - threshold (float): minimum similarity accepted by the agent
    Output:
        - Updated city grid (list of lists)
    '''
    
    similarity_score = similarity(grid, location)

    # Case 1: agent is satisfied
    if similarity_score >= threshold:
        return grid

    else:
        city_clone = clone_grid(grid)
        available = list_available(grid, location, threshold)
        i, j = location
        group = grid[i][j]

        # Case 2: no available spots
        if len(available)==0:
            return grid
        # Case 3: only one available spot
        elif len(available)==1:
            best_location = available[0]
        # Case 4: more than one available spot
        else:
            for n, new_location in enumerate(available):
                distance = euclidean_distance(location, new_location)
                # if location is the first of list, we assign it as the best
                if n == 0:
                    best_location = new_location
                    best_distance = distance
                # if not, we update best candidate only if location is closer
                else:
                    if distance < best_distance:
                        best_location = new_location
                        best_distance = distance

        new_i, new_j = best_location
        city_clone[new_i][new_j] = group
        city_clone[i][j] = ''
        return city_clone
```

## Task 5: Simulate one round of the model



```python
def one_round(grid, threshold):
    
    '''
    Simulates one round of the model in a city grid given a minimum similarity threshold
    
    Inputs:
        - city grid (list of lists): total grid
        - threshold (float): minimum similarity accepted by an agent
    Outputs:
        - A tuple with two items:
            - First item is the city grid at the end of the round
            - Second item is the number of moves in the round
    '''
    
    N = len(grid)
    city_clone = clone_grid(grid)
    moves = 0 # move counter

    # iterating through rows
    for row_i in range(N):
        # iterating through columns
        for col_i in range(N):
            # checking that location is not empty
            if city_clone[row_i][col_i] != '':
                new_city = move_agent(city_clone, [row_i, col_i], threshold)
                # now we check if attempting to move the agent produced a
                # change in the grid. We only add +1 to the move counter if
                # the resulting grid is different to the initial one
                if new_city != city_clone:
                    moves += 1
                    # updating the old grid to compare results
                    # in the next iteration
                    city_clone = new_city

    return new_city, moves
```

## Task 6: Running the model



```python
def run_schelling(grid, threshold, max_rounds):
    
    '''
    A function that runs our implementation of Schelling's model of segregation
    
    Inputs:
        - city grid (list of lists): total grid
        - threshold (float): minimum similarity accepted by an agent
    Outputs:
        - A tuple with two items:
            - first item is the final grid at the end of the simulation
            - second item is the number of rounds that passed in the simulation
    '''
    
    moved = True
    rounds = 0 # rpunds counter
    city_clone = clone_grid(grid)

    while moved == True and rounds < max_rounds:

        city_clone, n_moves = one_round(city_clone, threshold)

        # if no one moved, we set moved to False and the loop will end
        if n_moves == 0:
            moved = False
        # if there were moves, we increase the round counter
        else:
            rounds += 1

    return city_clone, rounds
```
